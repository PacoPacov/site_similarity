# Eval corpus 2018 models with the new method

```python
corpus_2018_models = [
    ('combined_data_unweighted_lvl_one_128D.model', 0.641),
    ('combined_data_unweighted_lvl_one_64D.model', 0.647),
    ('combined_data_weighted_lvl_one_128D.model', 0.621),
    ('combined_data_weighted_lvl_one_64D.model', 0.62),
    ('corpus_2018_referral_sites_unweighted_128_D.model', 0.665),
    ('corpus_2018_referral_sites_unweighted_64_D.model', 0.651),
    ('corpus_2018_unweighted_lvl_one_128D.model', 0.651),
    ('corpus_2018_unweighted_lvl_one_64D.model', 0.638),
    ('corpus_2018_unweighted_lvl_three_128D.model', 0.656),
    ('corpus_2018_unweighted_lvl_three_256D.model', 0.66),
    ('corpus_2018_weighted_lvl_one_128D.model', 0.621),
    ('corpus_2018_weighted_lvl_one_64D.model', 0.619),
    ('corpus_2018_weighted_lvl_three_128D.model', 0.631),
    ('corpus_2018_weighted_lvl_three_256D.model', 0.639)
]

corpus_2018_result = []
for model, old_training_accuracy in corpus_2018_models:
    print(f'Using model: {model}')
    node2vec_model = load_node2vec_model(model)

    new_model_training_without_minmaxscaller = train_official_model(
        data_year='2018', node2vec_model=node2vec_model, normalize_features=False)  # without minmaxscaller
    new_model_training_without_minmaxscaller['Old Training Accuracy'] = old_training_accuracy
    new_model_training_without_minmaxscaller['Model Name'] = model

    corpus_2018_result.append(new_model_training_without_minmaxscaller)

columns = [
    'Model Name', 'Accuracy', 'Old Training Accuracy',
    'Balanced Accuracy score', 'F1 micro score',
    'F1 macro score', 'F1 weighted score',
    'MAE', 'Confusion matrix']

print(pd.DataFrame.from_records(corpus_2018_result)[columns])
```

# Check the new training models with the one we use already

```python
models = [
    ('corpus_2020_unweighted_lvl_three_128D.model', 0.64959),
    ('corpus_2020_unweighted_lvl_three_256D.model', 0.64959),
    ('corpus_2020_weighted_lvl_three_128D.model', 0.64959)
]

for model, old_training_accuracy in models:
    print(f'Using model: {model}')
    node2vec_model = load_node2vec_model(model)

    new_model_training_with_minmaxscaller = train_official_model(data_year='2020', node2vec_model=node2vec_model, normalize_features=True)  # with minmaxscaller
    new_model_training_with_minmaxscaller['old_training_accuracy'] = old_training_accuracy

    new_model_training_without_minmaxscaller = train_official_model(data_year='2020', node2vec_model=node2vec_model, normalize_features=False)  # without minmaxscaller
    new_model_training_without_minmaxscaller['old_training_accuracy'] = old_training_accuracy

    clf = SVC(decision_function_shape='ovo', probability=True, random_state=42)
    old_model_training = train_model(clf=clf, data_year='2020', node2vec_model=node2vec_model)
    old_model_training['old_training_accuracy'] = old_training_accuracy

    res = [
        new_model_training_with_minmaxscaller,
        new_model_training_without_minmaxscaller,
        old_model_training
    ]

    print(pd.DataFrame.from_records(res)[['Accuracy', 'Old Training Accuracy', 'Balanced Accuracy score',
                                          'F1 micro score', 'F1 macro score', 'F1 weighted score', 'MAE',
                                          'Confusion matrix']])
    print("*" * 20)
```
