```python
from utils.notebook_utils import load_corpus

corpus = load_corpus('new_corpus_2020.csv')

corpus_2020 = pd.DataFrame(corpus)

corpus_2020.head()
```

```python
corpus_2020_alexa_sections_disctribution = df[df['site'].isin(corpus_2020['source_url_processed'].values.tolist())]

corpus_2020_alexa_sections_disctribution.head()
```

```python
corpus_2020_alexa_sections_disctribution.shape
```

```python
from utils.notebook_utils import _ALEXA_SECTIONS_NAMES

populated = corpus_2020_alexa_sections_disctribution[list(_ALEXA_SECTIONS_NAMES.keys())].sum(axis=0)

populated_df = populated.reset_index().rename({'index': 'Alexa Rank Sections', 0:'Populated'}, axis='columns')
populated_df['NOT Populated'] = populated_df['Populated'].subtract(len(corpus_2020)).abs()
populated_df['Percentage Populated'] = populated_df['Populated'].div(len(corpus_2020))

populated_df[['Alexa Rank Sections', 'NOT Populated', 'Populated', 'Percentage Populated']].head(20)
```

```python
# process different subsections report

for section, subsections in _ALEXA_SECTIONS_NAMES.items():
    if subsections:
        for index, subsection in enumerate(subsections):
            normalized_subsection_name = subsection.replace('_', ' ')
            temp_df = corpus_2020_alexa_sections_disctribution[[f'{section}_{subsection}' for subsection in subsections]].sum(axis=0)
            temp_df = temp_df.reset_index().rename({'index': f'Alexa Rank {normalized_subsection_name} Sub-Sections', 0:'Sub-Section Population'}, axis='columns')
            temp_df['Sub-Section NOT Population'] = temp_df['Sub-Section Population'].subtract(len(corpus_2020)).abs()
            temp_df['Percentage Sub-Section Population'] = temp_df['Sub-Section Population'].div(len(corpus_2020))
            if index == len(subsections) - 1:
                print(temp_df[[f'Alexa Rank {normalized_subsection_name} Sub-Sections', 'Sub-Section NOT Population', 'Sub-Section Population', 'Percentage Sub-Section Population']])
        print('##########################################')
```
