```python
import json


def load_level_data(level=0):
    with open(f"/home/paco/Documents/site_similarity/data/scrapping_results/processed level {level} results.json") as f:
        data = json.load(f)

    print(f'Loaded level {level} data with lenght:{len(data)}')
    
    return data
```

```python
import pandas as pd
from stellargraph import StellarGraph
import networkx as nx
import matplotlib.pyplot as plt

def create_graph(lvl_data, root):
    edges = []
    for k in  lvl_data[root].keys():
        edges.append((root, k))
        for overlap_site in lvl_data[root][k]['score']:
            edges.append((k, overlap_site['url']))

    return pd.DataFrame(edges, columns=['source', 'target'])

plt.figure(num=None, figsize=(30, 28), dpi=50)
nx.draw_networkx(StellarGraph(edges=create_graph(lvl_one_data, 'fury.news')).to_networkx())
```
